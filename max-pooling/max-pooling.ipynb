# Max pooling

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("face.jpg")
```

```python
bw_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(bw_image, cmap="gray")
```

```python
image_width, image_height = bw_image.shape
```

```python
k_size = 10
```

```python
output = np.zeros(((image_width + 1) - k_size, (image_width + 1) - k_size))
```

```python
for row in range((image_width + 1) - k_size):
    for col in range((image_height-1) - k_size):
        output[row][col] = np.max(bw_image[row:row + k_size, col:col + k_size])
```

```python
plt.imshow(output, cmap="gray")
```
